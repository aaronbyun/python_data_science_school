+ **set operations**
 - 수학 연산과 동일
 - 교집합, 합집합, 차집합 등 지원

```python
a = {1, 2}
b = {2, 3, 4}

print(a.intersection(b))
print(a & b)

print(a.union(b))
print(a | b)

print(a.difference(b))
print(a - b)

print(a.symmetric_difference(b)) # 양쪽 모두에 들어있지 않은 멤버
print(a ^ b)

print(a.issubset(b))
print(a <= b) #부분 집합
```

* **과제** 
 1. 14577을 소인수 분해 하시오.
 2. 2^28와 근사값으로 계산한 2^28의 차이를 구하시오.
 3. 다음과 같이 출력되도록 구구단 프로그램을 작성하시오.
    
> 2 * 1 = 2

> 2 * 2 = 4

>    ...

>    3 * 1 = 3

>    ...

>    9 * 8 = 72

>    9 * 9 = 81


```python
num = 12
div = 2
factors = []

while num != 1:
    if num % div == 0:
        factors.append(div)
        num /= div
    else:
        div += 1
        
factors
```
