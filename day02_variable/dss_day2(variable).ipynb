* **comparison operator(비교 연산자)** 
  - 프로그래밍에서는 비교를 할 경우, = 대신 **==**를 사용

  - <=, >= (작거나 같음, 크거나 같음)

* **문자열 slicing**
  - 인덱스가 하나의 문자만을 추출한다면,
  - slicing은 부분 문자열을 추출한다고 볼 수 있음
  - [시작:끝]와 같이 명시하여 [시작, 끝)에 해당하는 부분 문자열을 추출
  - 시작, 끝 인덱스가 생략이 되어 있다면, 0부터 혹은 끝까지로 간주

````python
letters = '123456789abcd'

print(letters[0:1])
print(letters[1:1])
print(letters[1:13])
print(letters[3:])
print(letters[:4])
print(letters[:])

letters
````
