* **Lambda 함수**
 + 단일문으로 표현되는 익명함수
 + 익명함수란 이름이 없는 구현체만 존재하는 간단한 함수를 의미
 + 코드 상에서 한번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 1회성으로 만들어서 쓸 때 사용.

```python
def square(x): 
    return x ** 2

lambda x : x ** 2

lambda x : x + 2

square2 = lambda x : x ** 2

print(square)
print(square2)

print(square(4), square2(4))
```

```python
nums = [91, 29, 14, 72, 83, 45]

print(sorted(nums))

def digit(n):
    return n % 10

print(sorted(nums, key=digit))
print(sorted(nums, key=lambda x : x%10))
```

```python
nums = [(1, 2), (9, 5), (8, 4), (7, 6), (10, 2), (4, 5)]

def get_key(item):
    return item[1]

print(sorted(nums, key = get_key))
print(sorted(nums, key = lambda item : item[0], reverse = True))
```

* **filter, map, reduce**
 + lambda가 유용하게 사용되는 3가지 대표적 함수
 + 함수형 프로그래밍의 기본 요소이기도 함
 + filter : 특정 조건을 만족하는 요소만 남기고 필터링
 + map    : 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환
 + reduce : 차례대로 앞 2개의 원소를 가지고 연산. 연산의 결과가 또 다음 연산의 입력으로 진행됨. 따라서 마지막까지 진행되면 최종 출력은 한개의 값만 남게 됨

```python
nums = list(range(2, 100))

print([_f for _f in nums if _f])
print([x for x in nums if x % 2 == 0])
```

* **정상적으로 동작하는 코드**

```python
nums = [1, 2, 3]
pos = 2

try:
    print(nums[pos])
    print(nums[0])
except:
    print('error occurred')

print('hahaha again')
```

* **예외가 발생하는 코드**

```python
nums = [1, 2, 3]
pos = 4

try:
    print(nums[pos]) # index out of range 예외 발생
    print(1, 2, 3)
    print(4, 5, 6)
except:
    print('need a index between 0 and len(nums)-1', 'but got', pos)
    
print(nums[0])
```

* **finally**
 + 예외처리 코드에서 항상 마지막에 위치
 + 예외 발생 여부에 관계 없이 항상 실행해야할 코드

```python
nums = [1, 2, 3]
pos = 5

try:
    print(nums[pos])
except:
    print('need a index between 0 and len(nums)-1', 'but got', pos)
finally:
    print('task done')
    
print(nums[0])
```

* **collections module **
 - 리스트, 튜플, 딕셔너리, 셋을 제외한 다른 유용한 data structure를 제공하는 모듈
 - 아래의 링크에서 문서 확인
 https://docs.python.org/2/library/collections.html#module-collections

* Counter 
  - counting을 목적으로 하는 객체
* defaultdict
  - 기본값을 갖는 dictionary
* OrderedDict
  - 순서를 갖는 dictionary (key가 삽입된 순서)

```python
from collections import Counter

a = [1, 1, 2, 1, 2, 3, 3, 2, 4, 5, 5, 1]
b = Counter(a)
print(b)
print(b.most_common())
print(b[1])
print(b[5])
```

```python
from collections import defaultdict

a = [1, 1, 2, 1, 2, 3, 3, 2, 4, 5, 5, 1]

temp = {}
for i in a:
    if i in temp:
        temp[i] += 1
    else:
        temp[i] = 1


num_dict = defaultdict(int)
for i in a:
    num_dict[i] += 1
        
print(num_dict)
```

```python
from collections import OrderedDict

dict1 = OrderedDict()
dict1['aaa'] = 1
dict1['aac'] = 2
dict1['ab'] = 100
dict1['dd'] = 100

dict2 = {}
dict2['aaa'] = 1
dict2['aac'] = 2
dict2['ab'] = 100
dict2['dd'] = 100

print(dict1)
print(list(dict1.keys()))

print() 

print(dict2)
print(list(dict2.keys()))
```
